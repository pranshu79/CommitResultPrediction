# IMPORTING NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
```

# IMPORTING THE DATA

```python
ps2=pd.read_excel("E:/ML/CONVOLVE/EPOCH-2/PS2/FileAndResults_Problem2.xlsx")
ps2
```

# APPLYING NLP TO 1ST COLUMN OF THE DATASET

```python
# USING COUNT VECTORIZER
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
file_vect=vect.fit_transform(ps2["File Changed"])
```

```python
data=ps2.copy()
```

# USING ONEHOTENCODER TO CONVERT CATEGORICAL FEATURES

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

x=data.drop(["File Changed","Sanity Result"],axis=1)

categorical_features=['Component', 'Coder', 'Tester', 'Category','Sanity name', 'nets Tested']
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)])
transformed_x=transformer.fit_transform(x)
transformed_x
```

```python
# USING LABEL ENCODING TO MAKE THE LABELS NUMERICAL
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(data["Sanity Result"])
```

```python
# CONCATINATING VECTORS OF THE FIRST COLUMN (AFTER APPLYING NLP - COUNT VECTORIZER) WITH ONE HOT ENCODED DATA
# TO MAKE THE FINAL DATASET READY
transformed_x_df = pd.DataFrame(transformed_x.toarray())
file_vect_df = pd.DataFrame(file_vect.toarray())
d=pd.concat([transformed_x_df,file_vect_df],axis=1)
d
```

# APPLYING MODEL ON FINALLY PROCESSED DATA

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# SETTING UP RANDOM SEED
np.random.seed(11)

#SPLITTING DATA INTO TRAIN AND TEST SET
xtrain,xtest,ytrain,ytest=train_test_split(d,y,test_size=0.2)

#THE MODEL
model=RandomForestClassifier()

#FITTING THE DATA
model.fit(xtrain,ytrain)

```

```python
# TESTING THE MODEL ON TEST SET
model.score(xtest,ytest)
```

```python
# USING CROSS_VAL_SCORE FOR BETTER ANALYSIS OF PERFORMANCE OF THE MODEL

from sklearn.model_selection import cross_val_score
cross_val_score(model,d,y,cv=5),np.mean(cross_val_score(model,d,y,cv=5))
```

```python
# USING CLASSIFICATION REPORT FOR FURTHER ANALYSIS
from sklearn.metrics import classification_report
ypreds=model.predict(xtest)
rep=pd.DataFrame(classification_report(ytest,ypreds,output_dict=True))
rep
```

```python
# PRINTING THE CONFUSION METRICS
ypred=model.predict(xtest)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,ypred)
cm
```

```python
# TUNING HYPERPARAMETERS FOR IMPROVING THE MODEL

#MAKING A GRID OF 
grid={
    "n_estimators":[50,125,175],
    "max_depth":[None,5,10],
    "max_features":["auto","sqrt"],
    "min_samples_split":[2,4],
    "min_samples_leaf":[1,2]
}

#USING GRIDSEARCHCV
from sklearn.model_selection import GridSearchCV
gs_clf=GridSearchCV(estimator=model,param_grid=grid,cv=5,verbose=2)
gs_clf.fit(transformed_x,y)

# GETTING THE BEST HYPERPARAMETERS
gs_clf.best_params_
```

```python
np.random.seed(1)
model2=RandomForestClassifier(max_depth= None,
                                                 max_features= 'auto',
                                                 min_samples_leaf= 1,
                                                 min_samples_split= 2,
                                                 n_estimators= 50)
model2.fit(xtrain,ytrain) 
model2.score(xtest,ytest)
```
